# Automatização de Processos

## Problema passo a passo

- Passo 1: Buscar informações das ações de forma automática
- Passo 2: Criar as análises dos dados
- Passo 3: Enviar e-mail para o gestor com os resultados das análises

## Passo 1: Buscar informações das ações de forma automática

```python
import yfinance as yf
```

```python
ticker = str(input('Digite o código da ação: '))
dados = yf.Ticker(ticker).history('6mo')
```

```python
fechamento = dados.Close
```

```python
fechamento
```

```python
fechamento.plot()
```

## Passo 2: Criar as análises dos dados

- Cotação máxima
- Cotação mínima
- Cotação atual

```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)

print(f'A cotação máxima é: R$ {maxima}')
print(f'A cotação mínima é: R$ {minima}')
print(f'A cotação atual é: R$ {atual}')
```

## Passo 3: Enviar e-mail para o gestor com os resultados das análises

- Abrir nova aba no navegador
- Digitar endereço do gmail e pressionar Enter
- Clicar em "Escrever"
- Digitar o destinatário e pressionar Tab
- Digitar o assunto e pressionar Tab
- Digitar a mensagem do e-mail
- Clicar em "Enviar"

```python
import pyautogui
import pyperclip
```

```python
# Pause de 2 segundos entre os passos
pyautogui.PAUSE = 5

# Abrir nova aba no navegador
pyautogui.hotkey('ctrl', 't')

# Digitar endereço do gmail e pressionar Enter
pyperclip.copy('www.gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('enter')

# Clicar em "Escrever"
pyautogui.click(x=30, y=200)

# Digitar o destinatário e pressionar Tab
pyperclip.copy('exemplo@gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')

# Digitar o assunto e pressionar Tab
pyperclip.copy(f'Análise {ticker}')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')

# Digitar a mensagem do e-mail
mensagem = f'''
Prezado gestor,

conforme solicitado, segue análise da ação {ticker} referente aos últimos 6 meses:

Cotação máxima: R$ {maxima}
Cotação mínima: R$ {minima}
Cotação atual: R$ {atual}

Qualquer dúvida, estou à disposição.

Atenciosamente,
xxxxxxxxxx'''
pyperclip.copy(mensagem)
pyautogui.hotkey('ctrl', 'v')

# Clicar em "Enviar"
pyautogui.click(x=1062, y=970)
```
